```python
# ============================
# ENSEMBLE MODEL: LGBM + RNN (LAG 30) + FORECAST AGUSTUS–DESEMBER 2018
# ============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from lightgbm import LGBMRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from datetime import timedelta

# -------- 1. LOAD DATA ASLI UNTUK FORECASTING --------
listrik_all = pd.read_csv("AEP_hourly.csv")
listrik_all['Datetime'] = pd.to_datetime(listrik_all['Datetime'])
listrik_all.set_index('Datetime', inplace=True)
listrik_all = listrik_all[listrik_all.index.year >= 2005]
daily_full = listrik_all.resample('D').sum()

# -------- 2. SIAPKAN FITUR UNTUK TRAINING --------
daily = daily_full.copy()
daily['dayofweek'] = daily.index.dayofweek
daily['month'] = daily.index.month
daily['lag_1'] = daily['AEP_MW'].shift(1)
daily['lag_30'] = daily['AEP_MW'].shift(30)
daily['rolling_mean_7'] = daily['AEP_MW'].shift(1).rolling(7).mean()
daily['rolling_std_7'] = daily['AEP_MW'].shift(1).rolling(7).std()
daily.dropna(inplace=True)

# -------- 3. SPLIT TRAIN/TEST --------
train = daily[daily.index.year <= 2013]
test = daily[(daily.index.year >= 2014) & (daily.index.year <= 2017)]

X_train_lgbm = train.drop(columns=['AEP_MW'])
y_train = train['AEP_MW']
X_test_lgbm = test.drop(columns=['AEP_MW'])
y_test = test['AEP_MW']

# -------- 4. TRAIN LGBM --------
model_lgbm = LGBMRegressor()
model_lgbm.fit(X_train_lgbm, y_train)

# -------- 5. TRAIN RNN --------
scaler = MinMaxScaler()
scaled = scaler.fit_transform(daily[['AEP_MW']])

def create_sequences(data, window_size=30):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i])
        y.append(data[i])
    return np.array(X), np.array(y)

window_size = 30
X_all_rnn, y_all_rnn = create_sequences(scaled, window_size)
idx_start = len(train) - window_size
X_train_rnn = X_all_rnn[:idx_start]
y_train_rnn = y_all_rnn[:idx_start]

model_rnn = Sequential()
model_rnn.add(SimpleRNN(64, input_shape=(window_size, 1)))
model_rnn.add(Dense(1))
model_rnn.compile(optimizer='adam', loss='mse')
model_rnn.fit(X_train_rnn, y_train_rnn, epochs=10, batch_size=32, verbose=1)

# -------- 6. ENSEMBLE MODEL --------
y_pred_lgbm = model_lgbm.predict(X_test_lgbm)
X_test_rnn = X_all_rnn[idx_start:]
y_pred_rnn_scaled = model_rnn.predict(X_test_rnn)
y_pred_rnn = scaler.inverse_transform(y_pred_rnn_scaled)

min_len = min(len(y_pred_lgbm), len(y_pred_rnn))
X_meta = np.column_stack((y_pred_lgbm[-min_len:], y_pred_rnn.flatten()[-min_len:]))
y_meta = y_test[-min_len:]

meta_model = LinearRegression()
meta_model.fit(X_meta, y_meta)

# -------- 7. FORECAST AGUSTUS–DESEMBER 2018 --------
forecast_days = pd.date_range(start='2018-08-01', end='2018-12-31', freq='D')
last_known = daily.copy()
predictions = []

for day in forecast_days:
    row = {
        'dayofweek': day.dayofweek,
        'month': day.month,
        'lag_1': last_known['AEP_MW'][-1],
        'lag_30': last_known['AEP_MW'][-30] if len(last_known) >= 30 else last_known['AEP_MW'][-1],
        'rolling_mean_7': last_known['AEP_MW'][-7:].mean(),
        'rolling_std_7': last_known['AEP_MW'][-7:].std()
    }

    df_input = pd.DataFrame([row], index=[day])
    y_pred_lgbm_day = model_lgbm.predict(df_input)[0]

    last_seq = last_known['AEP_MW'][-window_size:].values.reshape(-1, 1)
    last_seq_scaled = scaler.transform(last_seq)
    input_rnn = last_seq_scaled.reshape((1, window_size, 1))
    y_pred_rnn_scaled = model_rnn.predict(input_rnn)
    y_pred_rnn_day = scaler.inverse_transform(y_pred_rnn_scaled)[0, 0]

    final_pred = meta_model.predict([[y_pred_lgbm_day, y_pred_rnn_day]])[0]
    predictions.append(final_pred)

    last_known.loc[day] = [final_pred, row['dayofweek'], row['month'],
                           row['lag_1'], row['lag_30'], row['rolling_mean_7'], row['rolling_std_7']]

forecast_df = pd.DataFrame({'AEP_MW': predictions}, index=forecast_days)
forecast_df.index.name = 'Datetime'

# Gabungkan dengan data aktual dari 2005–Juli 2018
data_actual = daily_full[daily_full.index < '2018-08-01']
daily_full = pd.concat([data_actual, forecast_df])
daily_full.index.name = 'Datetime'

# Plot hasil gabungan
plt.figure(figsize=(16,6))
plt.plot(daily_full.index, daily_full['AEP_MW'], label='Data Aktual', color='blue')
plt.plot(forecast_df.index, forecast_df['AEP_MW'], label='Forecast (Ags–Des 2018)', color='orange')
plt.axvline(pd.to_datetime('2018-08-01'), color='gray', linestyle='--', label='Awal Forecast')
plt.title('Prediksi Listrik Harian (Forecast 2018 Agustus–Desember)')
plt.xlabel('Tanggal')
plt.ylabel('AEP_MW')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# -------- 8. FORECAST 2019–2020 --------
forecast_2019_2020_days = pd.date_range(start='2019-01-01', end='2020-12-31', freq='D')
predictions_2019_2020 = []

for day in forecast_2019_2020_days:
    row = {
        'dayofweek': day.dayofweek,
        'month': day.month,
        'lag_1': last_known['AEP_MW'][-1],
        'lag_30': last_known['AEP_MW'][-30] if len(last_known) >= 30 else last_known['AEP_MW'][-1],
        'rolling_mean_7': last_known['AEP_MW'][-7:].mean(),
        'rolling_std_7': last_known['AEP_MW'][-7:].std()
    }

    df_input = pd.DataFrame([row], index=[day])
    y_pred_lgbm_day = model_lgbm.predict(df_input)[0]

    last_seq = last_known['AEP_MW'][-window_size:].values.reshape(-1, 1)
    last_seq_scaled = scaler.transform(last_seq)
    input_rnn = last_seq_scaled.reshape((1, window_size, 1))
    y_pred_rnn_scaled = model_rnn.predict(input_rnn)
    y_pred_rnn_day = scaler.inverse_transform(y_pred_rnn_scaled)[0, 0]

    final_pred = meta_model.predict([[y_pred_lgbm_day, y_pred_rnn_day]])[0]
    predictions_2019_2020.append(final_pred)

    last_known.loc[day] = [final_pred, row['dayofweek'], row['month'],
                           row['lag_1'], row['lag_30'], row['rolling_mean_7'], row['rolling_std_7']]

forecast_df_2019_2020 = pd.DataFrame({'AEP_MW': predictions_2019_2020}, index=forecast_2019_2020_days)
forecast_df_2019_2020.index.name = 'Datetime'

# Gabungkan semua ke satu data
final_combined = pd.concat([daily_full, forecast_df_2019_2020])
final_combined.index.name = 'Datetime'

# Simpan ke CSV
final_combined.to_csv("listrik_gabungan_2005_2020.csv", index_label='Datetime')

# Plot hasil
plt.figure(figsize=(16,6))
plt.plot(final_combined.index, final_combined['AEP_MW'], label='Data Aktual + Forecast', color='blue')
plt.plot(forecast_df_2018.index, forecast_df_2018['AEP_MW'], label='Forecast (Ags–Des 2018)', color='orange')
plt.plot(forecast_df_2019_2020.index, forecast_df_2019_2020['AEP_MW'], label='Forecast (2019–2020)', color='red')
plt.axvline(pd.to_datetime('2018-08-01'), color='gray', linestyle='--', label='Awal Forecast 2018')
plt.axvline(pd.to_datetime('2019-01-01'), color='black', linestyle='--', label='Awal Forecast 2019')
plt.title('Prediksi Listrik Harian (Gabungan Forecast 2018–2020)')
plt.xlabel('Tanggal')
plt.ylabel('AEP_MW')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
